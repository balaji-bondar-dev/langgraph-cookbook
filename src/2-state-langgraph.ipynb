```python
from typing import TypedDict

# create a state class
class MessageState(TypedDict):
    question: str
    answer: str
    reviewed_answer : str
```

```python
def generate_answer(state: MessageState)->dict:
    question = state["question"]
    answer = "I am in answer block"
    return {"answer" : answer}

def review_answer(state: MessageState)->dict:
    answer = state["answer"],
    reviewed_answer = "I am in review answer block"
    return {"reviewed_answer" : reviewed_answer}
```

```python
from langgraph.graph import StateGraph,START,END

builder = StateGraph(MessageState) #called with the state class

builder.add_node("generate_answer",generate_answer)
builder.add_node("review_answer",review_answer)

builder.add_edge(START,"generate_answer")
#builder.set_entry_point("generate_answer")
builder.add_edge("generate_answer","review_answer")
builder.add_edge("review_answer",END)

graph = builder.compile()
```

```python
result = graph.invoke({"question":"What is your name?"})
print(result)
```
