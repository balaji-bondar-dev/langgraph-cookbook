```python
from typing import TypedDict,Annotated
from langgraph.graph import StateGraph,START,END
import operator

# State class
class MessageState(TypedDict):
    input: str
    last_writer: str
    all_steps: Annotated[list[str],operator.add] # list+list
    counter: int
```

```python
def step_a(state:MessageState)->dict:
    print("❯❯❯❯ step_a running...")
    return {
        "last_writer":"step_a",
        "all_steps": ["step_a executed."],
        "counter": 1, 
    }

def step_b(state:MessageState)->dict:
    print("❯❯❯❯ step_b running...")
    return {
        "last_writer":"step_b",
        "all_steps": ["step_b executed."],
        "counter": state["counter"]+10, 
    }

def step_c(state:MessageState)->dict:
    print("❯❯❯❯ step_c running...")
    return {
        "last_writer":"step_c",
        "all_steps": ["step_c executed."],
        "counter": state["counter"]+100, 
    }

def show_result(state:MessageState)-> dict:
    print("❯❯❯❯ show_result running...")
    # print(f"last writer '{state['last_writer']}'")
    # print(f"all steps '{state['all_steps']}'")
    # print(f"counter '{state['counter']}'")
    return{}
```

```python
# Build graph
builder = StateGraph(MessageState)

builder.add_node("step_a", step_a)
builder.add_node("step_b", step_b)
builder.add_node("step_c", step_c)
builder.add_node("show_result", show_result)

# Define flow & add edges
builder.add_edge(START, "step_a")
builder.add_edge("step_a", "step_b")
builder.add_edge("step_b", "step_c")
builder.add_edge("step_c", "show_result")
builder.add_edge("show_result", END)

# Compile the graph
graph = builder.compile()
```

```python
result = graph.invoke(
    {
        "input": "Demo",
        "last_writer": "",
        "all_steps": [],
        "counter": 0,        
    }
)

print("❯❯❯❯ final state")
print(result)
```
