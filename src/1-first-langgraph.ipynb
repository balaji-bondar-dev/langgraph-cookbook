```python
from typing import TypedDict
from langgraph.graph import StateGraph,START,END

# Define shared state
class GraphState(TypedDict):
    question: str
    answer: str
    reviewed_answer: str
```

```python
# Node 1: generate answer
def generate_answer(state: GraphState):
    question = state["question"]
    answer = f"✎ᝰ.Answering Question: {question}"
    return {"answer": answer}

# Node 2: review answer
def review_answer(state: GraphState):
    answer = state["answer"]
    reviewed = answer + " ✔ (reviewed)"
    return {"reviewed_answer": reviewed}
```

```python
# Build graph
builder = StateGraph(GraphState)

builder.add_node("generate", generate_answer)
builder.add_node("review", review_answer)

# Define flow & add edges
builder.set_entry_point("generate")
builder.add_edge("generate", "review")
builder.add_edge("review", END)

# Compile the graph
graph = builder.compile()
```

```python

# Run the graph with a sample input
result = graph.invoke({
    "question": "What is LangGraph?"
})

print(result["reviewed_answer"])
```
